#Root finding
#newton method





```julia

function newton_solver(f, x0; tol = 1e-7, h = 1e-5)
    x = x0
    for i in 1:100
        df = ( f(x+h) - f(x-h) )/(2h)

        ddf = (f(x + h) - 2f(x) + f(x-h)) / (h^2)
        step = df/ddf

        x = x - step
        if abs(step) < tol
        return x
        end
    end
    
    return x
end

```
