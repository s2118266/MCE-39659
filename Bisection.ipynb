```julia
function bisec(f, a,b)
    maxiter = 500
    fa = f(a)
    fb = f(b)
    if f(a) * f(b) > 0
            print("There is no root")
    end

    
    for i in 1:maxiter
        if abs(fm) < tol || abs(b - a) < tol
            return m, i

        end
        m = (a + b)/2
        fm = f(m)
        

        if f(a) * f(m) <= 0
            b = m
            fb = fm
        else 
            a = m
            fa = fm

        end
    end

    return (a+b)/2, maxiter
end

            

        

        
```
